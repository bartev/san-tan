# Annotating Axes

## Annotating with Text with Box

```python

import numpy.random
import matplotlib.pyplot as plt

fig = plt.figure(1, figsize=(5,5))
fig.clf()

ax = fig.add_subplot(111)
ax.set_aspect(1)

x1 = -1 + numpy.random.randn(100)
y1 = -1 + numpy.random.randn(100)
x2 = 1. + numpy.random.randn(100)
y2 = 1. + numpy.random.randn(100)

ax.scatter(x1, y1, color="r")
ax.scatter(x2, y2, color="g")

bbox_props = dict(boxstyle="round", fc="w", ec="0.5", alpha=0.9)
ax.text(-2, -2, "Sample A", ha="center", va="center", size=20,
        bbox=bbox_props)
ax.text(2, 2, "Sample B", ha="center", va="center", size=20,
        bbox=bbox_props)


bbox_props = dict(boxstyle="rarrow", fc=(0.8,0.9,0.9), ec="b", lw=2)
t = ax.text(0, 0, "Direction", ha="center", va="center", rotation=45,
            size=15,
            bbox=bbox_props)

bb = t.get_bbox_patch()
bb.set_boxstyle("rarrow", pad=0.6)

ax.set_xlim(-4, 4)
ax.set_ylim(-4, 4)

plt.draw()
plt.show()
```

```python
fig = plt.figure(1, figsize = (10, 5))
fig.clf()
ax = fig.add_subplot(111)
ax.set_aspect(0.5)

x1 = -1. + numpy.random.randn(100)
y1 = -1. + numpy.random.randn(100)
x2 = 1. + numpy.random.randn(100)
y2 = 1. + numpy.random.randn(100)

plt.scatter(x1, y1, color = 'r')
plt.scatter(x2, y2, color = 'g')
bbox_props = dict(boxstyle = 'round', fc = 'w', ec = '0.5', alpha = 0.7)
ax.text(-2, -2, 
        'Sample A', 
        ha = 'center',
        va = 'center', 
        size = 20,
        bbox = bbox_props)
ax.text(2, 2, 
        'Sample B', 
        ha = 'center',
        va = 'center', 
        size = 20,
        bbox = bbox_props)

bbox_props = dict(boxstyle = 'rarrow',
                  fc = (0.8, 0.9, 0.9),
                  ec = 'b',
                  lw = 2)
t = ax.text(0, 0, 'Direction',
            ha = 'center', va = 'center',
            rotation = 45,
            size = 15,
            bbox = bbox_props)
bb = t.get_bbox_patch()
bb.set_boxstyle('rarrow', pad = 0.6)

ax.set_xlim(-4, 4)
ax.set_ylim(-4, 4)
plt.draw()
```

# Fancy box demo

```python
import matplotlib.patches as mpatch
import matplotlib.pyplot as plt
styles = mpatch.BoxStyle.get_styles()
styles
```

```python
for x, y in enumerate(sorted(styles.keys())):
    print 'x = {x}, y = {y}'.format(**locals())
```

```python
styles = mpatch.BoxStyle.get_styles()
spacing = 1.2

figheight = (spacing * len(styles) + .5)
fig1 = plt.figure(1, (4/1.5, figheight/1.5))
fontsize = 0.3 * 72

for i, stylename in enumerate(sorted(styles.keys())):
    fig1.text(0.5, (spacing * (float(len(styles)) - i) - 0.5)/figheight, stylename,
              ha="center",
              size=fontsize,
              transform=fig1.transFigure,
              bbox=dict(boxstyle=stylename, fc="w", ec="k"))
plt.draw()
plt.show()
```
